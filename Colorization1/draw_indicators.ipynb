```python
import matplotlib.pyplot as plt
import seaborn as sns
import re

# 设置 Seaborn 样式
sns.set_style('darkgrid')

# 假设你的日志文件路径是 "training_log.txt"
log_file_path = './Colorization1.log'

# 定义列表来保存提取的 epoch 和 loss
epochs = []
training_losses = []

# 读取日志文件并提取 epoch 和训练损失
with open(log_file_path, 'r') as f:
    for line in f:
        # 使用正则表达式提取 epoch 和训练损失
        match = re.search(r'Epoch: (\d+), Training loss: ([\d.]+)', line)
        if match:
            epoch = int(match.group(1))
            loss = float(match.group(2))
            epochs.append(epoch)
            training_losses.append(loss)

# 创建训练损失曲线图
plt.figure(figsize=(10, 5))
plt.plot(epochs, training_losses, color='r', label='Training Loss')

# 设置图表标题和标签
plt.title('Colorization1 Loss Curve', fontsize=14)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Training Loss', fontsize=12)
plt.legend()
plt.savefig('./Colorization1_Loss_Curve.png', dpi=500)
# 显示图表
plt.show()
```
