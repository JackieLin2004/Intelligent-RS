```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 图片文件路径
image_folder = './images/'
images = {
    'origin1': image_folder + 'origin1.jpg',
    'bicubic1': image_folder + 'bicubic1.jpg',
    'srresnet1': image_folder + 'srresnet1.jpg',
    'srgan1': image_folder + 'srgan1.jpg',
    'isrgan1': image_folder + 'isrgan1.jpg',
    'origin2': image_folder + 'origin2.jpg',
    'bicubic2': image_folder + 'bicubic2.jpg',
    'srresnet2': image_folder + 'srresnet2.jpg',
    'srgan2': image_folder + 'srgan2.jpg',
    'isrgan2': image_folder + 'isrgan2.jpg',
}

# 创建两行四列的子图布局
fig, axs = plt.subplots(2, 5, figsize=(15, 7))

plt.suptitle("Super Resolution Comparison", fontsize=12, fontweight='bold')

# 第一行图片：origin1, bicubic1, srresnet1, srgan1
axs[0, 0].imshow(mpimg.imread(images['origin1']))
axs[0, 0].set_title('Origin1')
axs[0, 1].imshow(mpimg.imread(images['bicubic1']))
axs[0, 1].set_title('Bicubic1')
axs[0, 2].imshow(mpimg.imread(images['srresnet1']))
axs[0, 2].set_title('SRResNet1')
axs[0, 3].imshow(mpimg.imread(images['srgan1']))
axs[0, 3].set_title('SRGAN1')
axs[0, 4].imshow(mpimg.imread(images['isrgan1']))
axs[0, 4].set_title('ISRGAN1')

# 第二行图片：origin2, bicubic2, srresnet2, srgan2
axs[1, 0].imshow(mpimg.imread(images['origin2']))
axs[1, 0].set_title('Origin2')
axs[1, 1].imshow(mpimg.imread(images['bicubic2']))
axs[1, 1].set_title('Bicubic2')
axs[1, 2].imshow(mpimg.imread(images['srresnet2']))
axs[1, 2].set_title('SRResNet2')
axs[1, 3].imshow(mpimg.imread(images['srgan2']))
axs[1, 3].set_title('SRGAN2')
axs[1, 4].imshow(mpimg.imread(images['isrgan2']))
axs[1, 4].set_title('ISRGAN2')

# 隐藏坐标轴
for ax in axs.flat:
    ax.axis('off')

# 调整布局
plt.tight_layout()

# 保存图像
plt.savefig('Super_Resolution_Comparison.png', dpi=1000)

# 显示图形
plt.show()
```
