```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 图片文件路径
image_folder = './colorization-images/'
images = {
    'gt1': image_folder + 'gt1.jpg',
    'gray1': image_folder + 'gray1.jpg',
    'c11': image_folder + 'c11.jpg',
    'c12': image_folder + 'c12.jpg',
    'gt2': image_folder + 'gt2.jpg',
    'gray2': image_folder + 'gray2.jpg',
    'c21': image_folder + 'c21.jpg',
    'c22': image_folder + 'c22.jpg',
}

# 创建两行四列的子图布局
fig, axs = plt.subplots(2, 4, figsize=(15, 9))

plt.suptitle("Colorization Comparison", fontsize=12, fontweight='bold')

# 第一行图片
axs[0, 0].imshow(mpimg.imread(images['gt1']))
axs[0, 0].set_title('Ground Truth 1')
axs[0, 1].imshow(mpimg.imread(images['gray1']))
axs[0, 1].set_title('Gray1')
axs[0, 2].imshow(mpimg.imread(images['c11']))
axs[0, 2].set_title('Colorization 11')
axs[0, 3].imshow(mpimg.imread(images['c21']))
axs[0, 3].set_title('Colorization 21')

# 第二行图片
axs[1, 0].imshow(mpimg.imread(images['gt2']))
axs[1, 0].set_title('Ground Truth 2')
axs[1, 1].imshow(mpimg.imread(images['gray2']))
axs[1, 1].set_title('Gray2')
axs[1, 2].imshow(mpimg.imread(images['c12']))
axs[1, 2].set_title('Colorization 12')
axs[1, 3].imshow(mpimg.imread(images['c22']))
axs[1, 3].set_title('Colorization 22')

# 隐藏坐标轴
for ax in axs.flat:
    ax.axis('off')

# 调整布局
plt.tight_layout()

# 保存图像
plt.savefig('Colorization_Comparison.png', dpi=1000)

# 显示图形
plt.show()
```
