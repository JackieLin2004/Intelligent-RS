```python
import re
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')

# 日志文件路径
log_file_path = './SRResNet.log'

# 初始化存储损失值的列表
loss_values = []

# 从日志文件中提取损失值
with open(log_file_path, 'r') as file:
    for line in file:
        match = re.search(r'Average Loss: ([\d.]+)', line)
        if match:
            loss_values.append(float(match.group(1)))

# 创建损失曲线
epochs = list(range(1, len(loss_values) + 1))

plt.figure(figsize=(10, 5))
plt.plot(epochs, loss_values, color='r', label='SRResNet Loss')
plt.title('SRResNet Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
# plt.grid()
plt.legend()
# plt.tight_layout()
plt.savefig('SRResNet_Loss_Curve.png', dpi=500)  # 保存损失曲线图
plt.show()  # 显示损失曲线图
```
